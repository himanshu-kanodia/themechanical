## 2) Create the below pattern using nested for loop in Python. 
 
 

```python
x=int(10)
k=2
for i in range(1,x):
    l=1
    if(i<6):
        while(l<=i):
            print("*",end=" ")
            l=l+1
        print("\n")
    j=i-k
    while((i>5) and (j>=1)):
        print("*",end=" ")
        j=j-1
    
    if(i>5):
        print("\n")
        k=k+2
```
